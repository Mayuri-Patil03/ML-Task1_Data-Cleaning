```python
#import libraries

import numpy as np
import pandas as pd
```

```python
#import dataset
ds=pd.read_csv("data.csv")
ds.info()
```

```python
ds.columns
```

```python
#handling missing value(using mean and scikit learn)
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values = np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])#double matrix for multiple values
```

```python
ds
```

```python
ds["Country"].unique()
```

```python
#Encoding categorical column(onehot encoding)
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder(sparse_output=False)#creat variable of OneHotEncoder 
encoded_data=one.fit_transform(ds[["Country"]])
print(encoded_data)
```

```python
#Creating dataframe
column_names=one.get_feature_names_out(["Country"])
encoded_df=pd.DataFrame(encoded_data,columns=column_names)
encoded_df
```

```python
#getting columns
column_names
```

```python
#joining original dataset and new dataframe
ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#drop actual categorical country column
ds.drop(columns=["Country"])
ds
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
#Feature Scaling(Reduce Large data values to its smaller datavalues)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
```

```python
#divide dataset into independent(input and dependent(output)variables
x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```

```python
#spliting dataset as Training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```
